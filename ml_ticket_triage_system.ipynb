**Translating the subject and body columns to English**



```python
from transformers import pipeline
from langdetect import detect
from bs4 import BeautifulSoup
import re

# Translation Model for Spanish, French, and Portuguese
translator_romance = pipeline(
    'translation',
    model='Helsinki-NLP/opus-mt-ROMANCE-en',
    max_length=1024,
    truncation=True
)

# Translation Model for German
translator_de = pipeline(
    'translation',
    model='Helsinki-NLP/opus-mt-de-en',
    max_length=1024,
    truncation=True
)

# Split text into chunks based on max tokens
def split_into_chunks(text, max_tokens, tokenizer):
    tokens = tokenizer.tokenize(text)
    chunks = [tokens[i:i + max_tokens] for i in range(0, len(tokens), max_tokens)]
    return [tokenizer.convert_tokens_to_string(chunk) for chunk in chunks]

# Function to detect language and translate
def translate_text_dynamic(text):
    if not isinstance(text, str):
        return text

    try:
        # Remove HTML tags in the text
        text = BeautifulSoup(text, "html.parser").get_text()

        # Remove escape sequences (newlines, tabs, etc.)
        text = re.sub(r'[\n\r\t\\\f\v]', ' ', text)

        # Detect language
        lang = detect(text)

        # Choosing the right model based on language
        if lang in ['es', 'fr', 'pt']:
            translator = translator_romance
        elif lang == 'de':
            translator = translator_de
        else:
            return text

        # Split into chunks
        tokenizer = translator.tokenizer
        chunks = split_into_chunks(text, max_tokens=512, tokenizer=tokenizer)

        # Translate each chunk and combine results
        translated_chunks = [translator(chunk)[0]['translation_text'] for chunk in chunks]

        return " ".join(translated_chunks)

    except Exception as e:
        print(f"Error processing text: {text[:200]}... with error: {e}")
        return f"Error: {e}"
```

```python
# Apply translation dynamically to subject
df['subject_translated'] = df['subject'].apply(translate_text_dynamic)

# Display the translated dataframe
df[['subject','subject_translated']]
```

```python
# Apply translation dynamically to body
df['body_translated'] = df['body'].apply(translate_text_dynamic)

# Display the translated dataframe
df[['body','body_translated']]
```

```python
# Importing library for text data cleaning
import nltk
import contractions
from textblob import TextBlob
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
```

```python
# Function to correct spelling
def correct_spelling(text):
    # Ensure text is a string
    if isinstance(text, str):
        # Convert corrected TextBlob output to string
        return str(TextBlob(text).correct())r
    return text
```

```python
# Standardizing categories to lowercase
df['queue'] = df['queue'].str.lower()
df['priority'] = df['priority'].str.lower()
df['tags'] = df['tags'].str.lower()
df['subject'] = df['subject'].str.lower()
df['body'] = df['body'].str.lower()

# Removing leading/trailing whitespace
df['tags'] = df['tags'].str.strip()
df['subject'] = df['subject'].str.strip()
df['body'] = df['body'].str.strip()

# Expanding common contractions
df['subject'] = df['subject'].apply(lambda x: contractions.fix(x))
df['body'] = df['body'].apply(lambda x: contractions.fix(x))

# Apply spell correction to the 'subject' and 'body' columns
df['subject'] = df['subject'].apply(correct_spelling)
df['body'] = df['body'].apply(correct_spelling)

# Removing URLs
df['subject'] = df['subject'].apply(lambda x: re.sub(r'http\S+|www\S+', '', x))
df['body'] = df['body'].apply(lambda x: re.sub(r'http\S+|www\S+', '', x))

# Remove email addresses
df['subject'] = df['subject'].apply(lambda x: re.sub(r'\S+@\S+', '', x))
df['body'] = df['body'].apply(lambda x: re.sub(r'\S+@\S+', '', x))
```

**NLP Tokenization - subject body and tags column**

```python
# Tokenizing words for tags, subject and body
df['subject_word_tokenize'] = df['subject'].apply(nltk.word_tokenize)
df['body_word_tokenize'] = df['body'].apply(nltk.word_tokenize)
df['tags_word_tokenize'] = df['tags'].apply(nltk.word_tokenize)
```

```python
# Dropping tags, subject columns after tokenization
df = df.drop(['subject', 'tags'], axis=1)
```

```python
# Removing stopwords in word tokenization
stop_words = set(stopwords.words('english'))
df['tags_word_tokenize'] = df['tags_word_tokenize'].apply(lambda words: [word for word in words if word not in stop_words])
df['subject_word_tokenize'] = df['subject_word_tokenize'].apply(lambda words: [word for word in words if word not in stop_words])
df['body_word_tokenize'] = df['body_word_tokenize'].apply(lambda words: [word for word in words if word not in stop_words])
```

```python
# Removing punctuation in word tokenization
df['tags_word_tokenize'] = df['tags_word_tokenize'].apply(lambda words: [word for word in words if re.match(r'^\w+$', word)])
df['subject_word_tokenize'] = df['subject_word_tokenize'].apply(lambda words: [word for word in words if re.match(r'^\w+$', word)])
df['body_word_tokenize'] = df['body_word_tokenize'].apply(lambda words: [word for word in words if re.match(r'^\w+$', word)])
```

**Stemming - subject_word_tokenize and tags_word_tokenize columns**

```python
# Initialize the stemmer
stemmer = PorterStemmer()

# Function to stemming a list of words
def stem_words(words):
    return [stemmer.stem(word) for word in words]

# Applying stemming function to the subject_word_tokenize and tags_word_tokenize columns
df['subject_word_tokenize'] = df['subject_word_tokenize'].apply(stem_words)
df['tags_word_tokenize'] = df['tags_word_tokenize'].apply(stem_words)
```

**Lemmatization - body_word_tokenize**

```python
# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to lemmatize a list of words
def lemmatize_words(words):
    return [lemmatizer.lemmatize(word) for word in words]

# Applying lemmatize_words function to the body_word_tokenize column
df['body_word_tokenize'] = df['body_word_tokenize'].apply(lemmatize_words)
```

```python
# Listing first 10 rows of the new dataframe of the customer support tickets
df.head(10)
```

**Bag of Words - tags column**

```python
# Importing library to apply bag of words method
from sklearn.feature_extraction.text import CountVectorizer

# Converting the token lists into strings for CountVectorizer
df['tags_string'] = df['tags_word_tokenize'].apply(lambda x: ' '.join(x))

# Creating a CountVectorizer instance
bow_vectorizer = CountVectorizer()
# Applying the CountVectorizer to the tags_string
bow_matrix = bow_vectorizer.fit_transform(df['tags_string'])

# Convert bow_matrix matrix to a DataFrame
tags_bow_df = pd.DataFrame(bow_matrix.toarray(), columns=bow_vectorizer.get_feature_names_out())
tags_bow_df = tags_bow_df.add_prefix('tags_')

# Display the tags_bow_df DataFrame
tags_bow_df
```

**TF-IDF - subject and body column**

```python
# Importing library to apply TF-IDF method
from sklearn.feature_extraction.text import TfidfVectorizer

# Converting the subject word lists into strings for TfidfVectorizer
df['subject_string'] = df['subject_word_tokenize'].apply(lambda x: ' '.join(x))

# Creating a TfidfVectorizer instance
subject_vectorizer = TfidfVectorizer()
# Applying the TfidfVectorizer to the subject_string
subject_tfidf = subject_vectorizer.fit_transform(df['subject_string'])

# Convert TF-IDF matrices to DataFrames
subject_tfidf_df = pd.DataFrame(subject_tfidf.toarray(), columns=subject_vectorizer.get_feature_names_out())
subject_tfidf_df = subject_tfidf_df.add_prefix('subject_')

# Display the subject_tfidf_df DataFrame
subject_tfidf_df
```

```python
# Converting the body word lists into strings for TfidfVectorizer
df['body_string'] = df['body_word_tokenize'].apply(lambda x: ' '.join(x))

# Creating a TfidfVectorizer instance
body_vectorizer = TfidfVectorizer()
# Applying the TfidfVectorizer to the body_string
body_tfidf = body_vectorizer.fit_transform(df['body_string'])

# Convert TF-IDF matrices to DataFrames
body_tfidf_df = pd.DataFrame(body_tfidf.toarray(), columns=body_vectorizer.get_feature_names_out())
body_tfidf_df = body_tfidf_df.add_prefix('body_')

# Display the subject_tfidf_df DataFrame
body_tfidf_df
```

# Cross Validation Report Function

**Function for cross-validation performance evaluation for multiple models**

```python
def cross_validation_report(models, X_train, y_train):
    # Initialize Stratified K-Fold cross-validation (
    kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    # Define evaluation metrics with macro averaging for multi-classification
    scorers = {
        "Accuracy": "accuracy",
        "Precision": make_scorer(precision_score, average="macro"),
        "Recall": make_scorer(recall_score, average="macro"),
        "F1 Score": make_scorer(f1_score, average="macro")
    }

    # Empty list to store cross-validation results
    results = []

    # Loop over each model and each scoring metric
    for model_name, model in models.items():
        for metric_name, scorer in scorers.items():
            # Perform cross-validation and compute scores
            scores = cross_val_score(model, X_train, y_train, cv=kf, scoring=scorer)
            # Append the mean and standard deviation scores
            results.append({
                "model": model_name,
                "metric": metric_name,
                "score": scores.mean(),
                "std_dev": scores.std()
            })

    # Creating the dataframe based on the results
    df_results = pd.DataFrame(results)

    # Returning dataframe
    return df_results
```

# Model Speed and Accuracy Comparison Function




**Function to evaluate multiple model speed and accuracy**

```python
def model_speed_and_accuracy_comparison(models, X_train, X_test, y_train, y_test):

    # Empty list to store performance results for each model
    results = []

    # Looping over each model for its performance
    for model_name, model in models.items():
        # Record training start time
        start_train = time.time()
        # Training the model
        model.fit(X_train, y_train)
        # Calculating training duration
        train_time = time.time() - start_train

        # Record prediction start time
        start_pred = time.time()
        # Making predictions
        y_pred = model.predict(X_test)
        # Calculating prediction duration
        pred_time = time.time() - start_pred

        # Calculate accuracy score
        accuracy = accuracy_score(y_test, y_pred)

        # Append the results for the model
        results.append({
            "Model": model_name,
            "Training Time (s)": train_time,
            "Prediction Time (s)": pred_time,
            "Accuracy": accuracy
        })

    # Return the results as a DataFrame
    return pd.DataFrame(results)
```

# Visualisation Comparison Functions

**Function to create bar chart with error bars for multiple model based on cross validation results**



```python
def bar_chart_model_comparison(df, title):

  # Create a pivot table of scores with models as rows and metrics as columns
  pivot_scores = df.pivot(index='model', columns='metric', values='score')

  # Create a pivot table of standard deviations (errors) as rows and metrics as columns
  pivot_errors = df.pivot(index='model', columns='metric', values='std_dev')

  # List of model names
  models = pivot_scores.index.tolist()
  # List of evaluation metrics
  metrics = pivot_scores.columns.tolist()
  # Numeric x-axis positions for each model
  x = np.arange(len(models))
  # Width of each bar in the bar chart
  width = 0.2

  # Creating a figure and axis for plotting
  fig, ax = plt.subplots(figsize=(12, 6))

  # Adding a grid to the plot for readability
  plt.grid(True, linestyle=':', linewidth=1.2, alpha=0.7)

  # Create grouped bars with error bars for each metric
  for i, metric in enumerate(metrics):
      scores = pivot_scores[metric].values
      errors = pivot_errors[metric].values
      ax.bar(x + (i - len(metrics) / 2) * width, scores, width, yerr=errors, capsize=4, label=metric)


  # Dynamic title name
  title_name = title

  # Setting title for the bar chart with error bars
  ax.set_title(title_name, pad=10)

  # Label for the y-axis
  ax.set_ylabel('Metric Score')
  # Set y-axis limits
  ax.set_ylim(0.6, 1.00)

  # Set x-tick positions
  ax.set_xticks(x)
  # Label x-ticks with model names
  ax.set_xticklabels(models)

  # Move legend outside the plot to the right
  ax.legend(title='Metric', bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0.)

  # Adjust layout of the plot
  plt.tight_layout(rect=[0, 0, 0.85, 1])

  # Saving the plot
  plt.savefig(f"/content/drive/MyDrive/Colab Notebooks/screenshots/{title}.jpg")

  # Display the plot
  plt.show()
```

**Function to create learning curve for multiple models**

```python
def learning_curves_model_comparison(models, X_train, y_train, title):

    # Initialize Stratified K-Fold cross-validation
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    # Define a range of training set sizes to evaluate
    train_sizes = np.linspace(0.1, 1.0, 10)

    # Extracting model names and model instances
    model_names = list(models.keys())
    model_objects = list(models.values())

    # Create a 3x2 grid of subplots to plot the learning curves
    fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 16))
    axes = axes.flatten()

    # Looping through each model and generating learning curve
    for i, (name, model) in enumerate(zip(model_names, model_objects)):

        # Computing learning curve data using cross-validation
        train_sizes_out, train_scores, validation_scores = learning_curve(
            estimator = model,
            X = X_train,
            y = y_train,
            train_sizes = train_sizes,
            cv = cv,
            scoring = 'accuracy',
            n_jobs = -1,
            error_score = 'raise'
        )

        # Calculate mean accuracy scores across folds
        train_scores_mean = np.mean(train_scores, axis=1)
        validation_scores_mean = np.mean(validation_scores, axis=1)

        # Plot training and validation accuracy curves
        axes[i].plot(train_sizes_out, train_scores_mean, linestyle='--', label='Training Accuracy')
        axes[i].plot(train_sizes_out, validation_scores_mean, label='Validation Accuracy')

        # Add plot title and formatting
        axes[i].set_title(f"{name} Learning Curve", pad=15, fontweight="bold")

        # Label only left column with "Accuracy"
        if i % 2 == 0:
            axes[i].set_ylabel("Accuracy")
        else:
            axes[i].set_ylabel("")

        # Label only bottom row with "Training Size"
        if i >= len(models) - 2:
            axes[i].set_xlabel("Training Size")
        else:
            axes[i].set_xlabel("")

        # Add legend and grid
        axes[i].legend(loc='lower right')
        axes[i].grid(True)

    # Remove any unused subplots (hiding 6th subplot)
    for j in range(len(models), len(axes)):
        fig.delaxes(axes[j])

    # Add a main title for the entire figure
    plt.suptitle(f"{title}", fontsize=18, fontweight="bold")

    # Adjust layout to accommodate spacing for the main title
    plt.tight_layout(rect=[0, 0.07, 1, 0.98])

    # Saving the plot
    plt.savefig(f"/content/drive/MyDrive/Colab Notebooks/screenshots/{title}.jpg")

    # Display the full plot
    plt.show()
```

**Function to create confusion matrix for multiple modes**

```python
def confusion_matrices_model_comparison(models, X_train, y_train, X_test, y_test, label_names, title):

    # Creating 3x2 grid of subplots for displaying the confusion matrix
    fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 16))
    axes = axes.flatten()


    # Loop over each model and its name
    for i, (name, model) in enumerate(models.items()):
        # Training the model using the training data
        model.fit(X_train, y_train)
        # Generate predictions on the test set
        y_pred = model.predict(X_test)
        # Computing the confusion matrix
        cm = confusion_matrix(y_test, y_pred)

        # Plot the confusion matrix using a heatmap
        sns.heatmap(cm, annot=True, fmt='d',cmap="coolwarm",ax=axes[i])
        # Setting the title of confusion matrix based on the model
        axes[i].set_title(f"{name} Confusion Matrix", pad=15, fontweight="bold")

        # Label only left column with "True Labels"
        if i % 2 == 0:
            axes[i].set_ylabel("True Labels")
        else:
            axes[i].set_ylabel("")

        # Label only bottom row with "Predicted Labels"
        if i >= len(models) - 2:
            axes[i].set_xlabel("Predicted Labels")
        else:
            axes[i].set_xlabel("")

    # Remove any unused subplots (hiding 6th subplot)
    for j in range(len(models), len(axes)):
        fig.delaxes(axes[j])

    # Add a main title for the entire figure
    plt.suptitle(title, fontsize=16, fontweight="bold", y=1)

    # Adjust layout to avoid overlap
    plt.tight_layout()

    # Creating the label legend from label names
    label_legend = "Target Labels:  [" + ", ".join(f"{i}: {label}" for i, label in enumerate(label_names)) + "]"
    # Displaying the label legend below the plot
    plt.figtext(0.5, -0.02, label_legend, ha='center', fontsize=11, wrap=True, bbox=dict(facecolor='whitesmoke', edgecolor='gray', boxstyle='round,pad=0.5'))

    # Saving the plot
    plt.savefig(f"/content/drive/MyDrive/Colab Notebooks/screenshots/{title}.jpg", bbox_inches='tight')

    # Display the plot
    plt.show()
```

**Reading the dataframe of translated customer tickets**

```python
# Reading translated dataframe of customer tickets
df_translated = pd.read_csv('drive/MyDrive/Colab Notebooks/data/helpdesk_customer_tickets_translated.csv')
```

**Applying text pre-processing on the translated body column**

```python
# Create a new DataFrame for text summarization containing only the 'body' column
df_text_summarization = df_translated[['body']].copy()

def preprocess_text(text):
    # Expand contractions
    text = contractions.fix(text)
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove URLs (http, https, www)
    text = re.sub(r'http\S+|www\S+', '', text)
    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)
    # Replace multiple whitespace characters with a single space
    text = re.sub(r'\s+', ' ', text)
    # Remove leading and trailing whitespace
    text = text.strip()

    return text

# Apply the preprocessing function to the 'body' column of the DataFrame
df_text_summarization['body_cleaned'] = df_text_summarization['body'].apply(preprocess_text)
```

```python
def remove_greeting_and_sign_off(text):

    # Define regex patterns for greetings and sign-offs
    patterns = [
        r'^(dear|hello|hi|hey|greetings|good (morning|afternoon|evening))[\s,]*',
        r'(sincerely|thanks|thank you|regards|best regards|kind regards|warm regards)[\s\S]*$',
    ]

    # Apply each pattern to remove matching text
    for pattern in patterns:
        text = re.sub(pattern, '', text, flags=re.IGNORECASE)

    # return text after removing greeting and sign off text
    return text.strip()

# Apply the preprocessing function to the 'body' column of the DataFrame
df_text_summarization['body_cleaned'] = df_text_summarization['body'].apply(remove_greeting_and_sign_off)
```

**Reading the summarization_text_dataset**

```python
# Reading the dataset for text summarization.
df_text_summarization = pd.read_csv('drive/MyDrive/Colab Notebooks/data/summarization_text_dataset.csv')
```

**Initialize models for extractive and abstractive summarization**

```python
# Loading BERT for extractive summarization
bert_summarizer_model = Summarizer()

# Loading BART for abstractive summarization
bart_summarizer_model = pipeline("summarization", model="facebook/bart-large-cnn")
```

**BERT and BART summarization functions**

```python
# Function to summarize text using a BERT-based summarization model
def bert_summarize(text):

    try:
        # Count the number of words in the input text
        word_count = len(text.split())

        # If the text is too short skip summarization
        if word_count < 25:
            return text.strip().capitalize()

        # Generate summary using BERT summarizer
        summary = bert_summarizer_model(text, ratio=0.7)

        # Return the summarized text
        return summary


    # Return a error message if an exception occurs
    except Exception as e:
        return f"[Error]: {e}"

# Function to summarize text using a BART-based summarization model
def bart_summarize(text):

    try:
        # Count the number of words in the input text
        word_count = len(text.split())

        # If the text is too short skip summarization
        if word_count < 25:
            return text.strip().capitalize()

        # Calculate max and min length for the summary based on input size
        max_len = max(20, int(word_count * 0.7))
        min_len = max(10, int(word_count * 0.3))

        # Generate summary using BART summarizer
        output = bart_summarizer_model(
            text,
            max_length=max_len,
            min_length=min_len,
            do_sample=False
        )

        # Extract the summary text from the model output
        summary = output[0]['summary_text']

        # Return the summarized text
        return summary.strip().capitalize()

    # Return a error message if an exception occurs
    except Exception as e:
        return f"[Error]: {e}"
```

**Applying extractive and abstractive summarization**

```python
# Applying BERT summarization to the 'body' column
df_text_summarization["summary_extractive"] = df_text_summarization["body_cleaned"].apply(bert_summarize)

# Applying BART summarization to the 'body' column
df_text_summarization["summary_abstractive"] = df_text_summarization["body_cleaned"].apply(bart_summarize)

# Dropping the "body_cleaned" column from the dataframe
df_text_summarization = df_text_summarization.drop("body_cleaned", axis=1)

# Saving the results as a CSV file for the capstone project
df_text_summarization.to_csv('drive/MyDrive/Colab Notebooks/data/summarization_text_results.csv', index=False)
```
